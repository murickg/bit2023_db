```python
r = redis.Redis(host='localhost', port=6379, decode_responses=True)
```

Store and retrieve a simple string

```python
r.set('foo', 'bar')
```

```python
r.get('foo')
```

Вроде как Redis работает

```python
# Open the file 'News_Category_Dataset.json' in read-only mode and assign its object to the 'file' variable.
file = open('News_Category_Dataset.json', 'r')

# Read the contents of the file and assign it to a string variable 'str'.
str = file.read()

# Close the file object to free up system resources.
file.close()

# Slice the string to extract only the first 1,000,000 characters and reassign it to 'str' variable.
str = str[:10**6]
```

```python
r.set('files', str)
```

```python
r.get('files')
```

```python
d = {i: i for i in range(10**6)}
```

```python
r.hset('dict', mapping=d)
```

```python
r.hvals('dict')
```

```python
r.zadd('zadd', mapping=d)
```

```python
r.zcard('zadd')
```

```python
r.zcount('zadd', 300, 70000)
```

```python
for i in range(10**4):
    r.lpush('list', i)
```

```python
for i in range(10**4):
    r.lpop('list', i)
```
